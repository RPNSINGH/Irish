# How to visulise Desion tree graph made by our decision tree.

```python
import time
import random
from tkinter import messagebox
from tkinter import *
import pandas as pd
from sklearn.datasets import load_iris


ir=load_iris()
df=pd.DataFrame(ir.data,columns=ir.feature_names)
###############       ALGORITHMS      ########################################
def knn(sl,sw,pl,pw):
    from sklearn.neighbors import KNeighborsClassifier
    knn = KNeighborsClassifier(n_neighbors=3)
    knn.fit(df,ir.target)
    pred=knn.predict([[sl,sw,pl,pw]])
    return pred
def svm(sl,sw,pl,pw):
    from sklearn.svm import SVC
    sv=SVC()
    sv.fit(df,ir.target)
    pred=sv.predict([[sl,sw,pl,pw]])
    return pred
def dt(sl,sw,pl,pw):
    from sklearn.tree import DecisionTreeClassifier
    dt=DecisionTreeClassifier()
    dt.fit(df,ir.target)
    pred=dt.predict([[sl,sw,pl,pw]])
    return pred
def rf(sl,sw,pl,pw):
    from sklearn.ensemble import RandomForestClassifier
    rfc=RandomForestClassifier()
    rfc.fit(df,ir.target)
    pred=dt.predict([[sl,sw,pl,pw]])
    return pred


m=Tk()
m.geometry("1000x600")
m.title("Iris_Flower")
#m.configure(background="#D7BDE2")
######################  variables     ##################
sl=DoubleVar()
sw=DoubleVar()
pl=DoubleVar()
pw=DoubleVar()
####    FRAMES  ###############################################
top=Frame(m,height=50,width=1000,relief=SUNKEN)
top.pack(side=TOP)
f1=Frame(m,height=600,width=600)
f1.pack(side=LEFT)
f2=Frame(m,height=600,width=400,relief=SUNKEN)
f2.pack(side=RIGHT)
f3=Frame(m,height=140,width=1000,relief=SUNKEN)
f3.pack(side=BOTTOM)
##############  result windows wrt algorithms        ##################
def knn_window():
    check=Toplevel()
    check.geometry("600x500+120+120")
    check.title("Approval_Status")
    l=Label(check,text="Flower Name Predicted by Knn :↓↓",font=('arial',16,'bold'), bd=10,anchor='nw').pack(side=TOP)
    button=Button(check,text="BACK",command=check.destroy,padx=4,pady=6,bd=6,fg='black',font=('arial',25,'bold'),\
          width=20).pack(side=BOTTOM)  
    t=Text(check,height=2,width=15,font=('arial',16,'bold'))
    t.pack()
    sl=e1.get()
    sw=e2.get()
    pl=e3.get()
    pw=e4.get()
    pred=knn(sl,sw,pl,pw)
    pred=list(pred)[0]
    pred=int(pred)
    if pred==0:
        flower="           Setosa   "
    elif pred==1:
        flower="          Versicolor    "
    else:
        flower="          Virginica    "
    t.insert(END,flower)
   
    canvas=Canvas(check,width=300,height=200)
    img=PhotoImage(file="I://flower//dt.png")
    canvas.create_image(0,0,anchor=NW,image=img)
    canvas.pack()
def sv_window():
    check=Toplevel()
    check.geometry("600x500+120+120")
    check.title("Approval_Status")
    l=Label(check,text="Flower Name Predicted by SVM :↓↓",font=('arial',16,'bold'), bd=10,anchor='nw').pack(side=TOP)
    button=Button(check,text="BACK",command=check.destroy,padx=4,pady=6,bd=6,fg='black',font=('arial',25,'bold'),\
          width=20).pack(side=BOTTOM)  
    t=Text(check,height=2,width=15,font=('arial',16,'bold'))
    t.pack()
    sl=e1.get()
    sw=e2.get()
    pl=e3.get()
    pw=e4.get()
    pred=svm(sl,sw,pl,pw)
    pred=list(pred)[0]
    pred=int(pred)
    if pred==0:
        flower="           Setosa   "
    elif pred==1:
        flower="          Versicolor    "
    else:
        flower="          Virginica    "
    t.insert(END,flower,)

def dt_window():
    check=Toplevel()
    check.geometry("600x500+120+120")
    check.title("Approval_Status")
    l=Label(check,text="Flower Name Predicted by Decision Tree :↓↓",font=('arial',16,'bold'), bd=10,anchor='nw').pack(side=TOP)
    button=Button(check,text="BACK",command=check.destroy,padx=4,pady=6,bd=6,fg='black',font=('arial',25,'bold'),\
          width=20).pack(side=BOTTOM)  
    t=Text(check,height=2,width=15,font=('arial',18,'bold'))
    t.pack()
    sl=e1.get()
    sw=e2.get()
    pl=e3.get()
    pw=e4.get()
    pred=dt(sl,sw,pl,pw)
    pred=list(pred)[0]
    pred=int(pred)
    if pred==0:
        flower="           Setosa   "
    elif pred==1:
        flower="          Versicolor    "
    else:
        flower="          Virginica    "
    t.insert(END,flower)

def rf_window():
    check=Toplevel()
    check.geometry("600x500+120+120")
    check.title("Approval_Status")
    l=Label(check,text="Flower Name Predicted by RFA :↓↓",font=('arial',16,'bold'), bd=10,anchor='nw').pack(side=TOP)
    button=Button(check,text="BACK",command=check.destroy,padx=4,pady=6,bd=6,fg='black',font=('arial',25,'bold'),\
          width=20).pack(side=BOTTOM)  
    t=Text(check,height=2,width=15,font=('arial',16,'bold'))
    t.pack()
    sl=e1.get()
    sw=e2.get()
    pl=e3.get()
    pw=e4.get()
    pred=dt(sl,sw,pl,pw)
    pred=list(pred)[0]
    pred=int(pred)
    if pred==0:
        flower="           Setosa   "
    elif pred==1:
        flower="          Versicolor    "
    else:
        flower="          Virginica    "
    t.insert(END,flower)
#********************   TIME    ***********************
#localtime=time.asctime(time.localtime(time.time()))
def get_time():
    time_string=time.strftime("%I:%M:%S %p")
    clock.config(text=time_string)
    clock.after(200,get_time)
clock=Label(top,font=("time",20,"bold"))
get_time() 
#********************  title and time   ***********************
lblinfo=Label(top,font=('arial',30,'bold'),text="USER'S INPUT",\
              fg='Steel Blue',anchor='w',bd=10)
lblinfo.grid(row=0)
clock.grid(row=1)

bsl=Label(f1,text="sepal length  :",font=('arial',14,'bold'),bd=10)
bsl.grid(row=0,column=0)
e1=Entry(f1,font=('arial',14,'bold'),bd=10,textvariable=sl)
e1.grid(row=0,column=1)

bsw=Label(f1,text="sepal width  :",font=('arial',14,'bold'),bd=10)
bsw.grid(row=1,column=0)
e2=Entry(f1,font=('arial',14,'bold'),bd=10,textvariable=sw)
e2.grid(row=1,column=1)

bpl=Label(f1,text="petal length  :",font=('arial',14,'bold'),bd=10)
bpl.grid(row=2,column=0)
e3=Entry(f1,font=('arial',14,'bold'),bd=10,textvariable=pl)
e3.grid(row=2,column=1)

bpw=Label(f1,text="petal width  :",font=('arial',14,'bold'),bd=10)
bpw.grid(row=3,column=0)
e4=Entry(f1,font=('arial',14,'bold'),bd=10,textvariable=pw)
e4.grid(row=3,column=1)


bg1=Button(f2,text="DT",width=7,height=8,font=('arial',14,'bold'),bd=10,bg="#D7BDE2",command=dt_window).grid(row=0,sticky="s")
bg2=Button(f2,text="RFA",width=7,height=8,font=('arial',14,'bold'),bd=10,bg="#64CEDC",command=rf_window).grid(row=0,column=1,sticky="s")
bg3=Button(f2,text="Knn",width=7,height=8,font=('arial',14,'bold'),bd=10,bg="#F9E79F",command=knn_window).grid(row=1,column=0,sticky="s")
def reset():
    sl.set('')
    sw.set('')
    pl.set('')
    pw.set('')
def call():
    mess = messagebox.askyesno("चेतावनी", "क्या आपको यकीन है ?")
    if (mess):
        m.destroy()
# def ext():
#     sw=messagebox.showwarning("चेतावनी", "Every entry will be cleared. ")
#     if (sw):
#         m.reset()
bg4=Button(f2,text="SVM",width=7,height=8,font=('arial',14,'bold'),bg="#E59866",bd=10,command=sv_window).grid(row=1,column=1,sticky="s")
reset=Button(f3,text="RESET",command=reset,width=15,height=3,font=('arial',15,'bold'),bd=6,bg="#CD6155").grid(row=0)
exit=Button(f3,text="EXIT",command=call,width=15,height=3,font=('arial',15,'bold'),bd=6,bg="#CD6155").grid(row=0,column=1)
m.mainloop()
```
